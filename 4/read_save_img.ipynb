```python
import cv2
import numpy as np

# 读取彩色图像
# img = cv2.imread('../lena.jpg', cv2.IMREAD_COLOR)
# 读取灰度图像
img = cv2.imread('../lena.jpg', cv2.IMREAD_GRAYSCALE)  # GRAY=0.3*R+0.59*G+0.11*B
# img_bgr = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR) #  单纯将灰度图在各通道赋值了三次
# 读取图像，包括图像原始的alpha通道
# img = cv2.imread('../lena.jpg', cv2.IMREAD_UNCHANGED)
print(np.array(img).shape)
```

```python
cv2.namedWindow("image", cv2.WINDOW_NORMAL)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('lena_gray.jpg', img)
```

```python
from matplotlib import pyplot as plt
import cv2

# 显示灰度图
img = cv2.imread('../lena.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
# plt.xticks([]), plt.yticks([])
plt.show()

# 使用matplot显示cv2读取的图片时，需先将cv2的bgr通道转换为rgb通道
img = cv2.imread('../lena.jpg', cv2.IMREAD_COLOR)
b,g,r = cv2.split(img)
img2 = cv2.merge([r,g,b])
plt.subplot(121);plt.imshow(img)
plt.subplot(122);plt.imshow(img2)
plt.show()
```
